### 普通最小二乘法，使用矩阵求逆的方法直接获得$\hat w$，基于：$\hat w=(X^T X)^{-1}X^T y$

#### 程序清单8-1 标准回归函数和数据导入函数

```python
import numpy as np


def loadDataSet(filename):
    numFeat=len(open(filename).readline().strip().split('\t'))-1
    dataMat=[]
    labelMat=[]
    for line in open(filename).readlines():
        lineArr=[]
        currentLine=line.strip().split('\t')
        for i in range(numFeat):
            lineArr.append(float(currentLine[i]))
        dataMat.append(lineArr)
        labelMat.append(float(currentLine[-1]))
    return dataMat,labelMat


dataPath='D:\\机器学习实战代码\\machinelearninginaction\\Ch08\\ex0.txt'
xArr,yArr=loadDataSet(dataPath)
print(np.mat(xArr).shape)
print(len(yArr))
```

### 基于：$\hat w=(X^T X)^{-1}X^T y$,计算$\hat w$

```python
def standRegres(xArr, yArr):
    # 转化为矩阵，注意y需要转为列向量
    xMat=np.mat(xArr)
    yMat=np.mat(yArr).T
    
    xTx=xMat.T*xMat
    if np.linalg.det(xTx)==0:
        print('xTx 不可逆！')
        return
    
    ws=xTx.I*(xMat.T*yMat)
    return ws


standRegres(xArr, yArr)
```
